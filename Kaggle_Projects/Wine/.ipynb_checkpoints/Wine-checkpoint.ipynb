```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
%matplotlib inline
```

```python
wine_data = pd.read_csv('winequality-red.csv')
```

```python
wine_data.head(20)
```

```python
wine_labels = wine_data['quality']
```

```python
sns.barplot(data=wine_data,x='quality',y='alcohol')
```

```python
label_q = LabelEncoder()
#bins = (2,4,6,8)
#groups = ['bad','good','very_good']
#lb = pd.cut(x = wine_labels,bins= bins,labels=groups)
wl = label_q.fit_transform(wine_labels)
```

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC(kernel=degree=20)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV,cross_validate
X = wine_data.drop('quality',axis=1)
y = wl
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
svm_model.fit(X_train,y_train)
```

```python
pred = svm_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
print (classification_report(y_true=y_test,y_pred=pred))
```

```python
print (accuracy_score(y_true=y_test,y_pred=pred))
```
